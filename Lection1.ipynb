# Про циклы, генераторы, лямбда выражения и не только
* ## цикл for
```
for i in range(10):
    print(i) 
# выведет последовательно значения от 0 до 9
for _ in range(10)
    pass 
# _ используется как заглушка, когда нам не нужно получать значение 
# pass используется для "бездействия", когда мы не хотим ничего делать 
a = ['a','b','c']
for c, i in enumerate(a) # если нужно пронумеровать что-либо
    print(f'{c} {i}') # последняя пара будет следующей: "2 'c'" 
    # не перепутайте нумерацию и "перечисление"!
b = ['d','e','f']
#сшивание списков для поочередного использования их элементов в списке
for i,j in zip(a,b):
    print(f'{i} {j}') # последняя пара: "'c' 'f'"  
```
* ## цикл while
```
# с while всё попроще: есть условие, пока его не достигнешь - исполняй 
i = 10
while i != 10:
    i +- 1 
```
* ## условия if,elif,else
```
#Используются для ветвления кода по условиям
a = True 
b = False
if a: # если тип a логический, то можно не писать if a == True 
    print(a)
# elif - это else if
elif not a # равноценное выражение: elif a == False
    print(a)
else:
    pass
```

```
class AutoMachine: # для имени класса используйте CamelCase
    def __init__(self,wheels,doors): #используется для инициализации переменных в классе
        self.wheels = wheels
        self.doors = doors
```
специальное слово "self" используется для возможности вызова переменных в методах класса. Класс-потомок может используя self вызывать методы класса-родителя. Давайте создадим класс-потомок марки машин, у которых есть название и страна-производитель 
```
class Brand(AutoMachine): # наследует класс AutoMAchine
    def __init__(self,wheels,doors,name,country):
        super.__init__(self,wheels,doors)
        # передает переменные в класс-родитель 
        self.name = name 
        self.country = country 

    def sayMyName(self): # метод класса Brand
        print(f'Im {self.name} from {self.country}')
```
Отлично! Теперь давайте создадм класс модели машины, у неё будет название модели и своя цена
```
class Model(Brand):
    def __init__(self,wheels,doors,name,country,model,price)
        super.__init__(self,wheels,doors,name,country)
        self.model = model 
        self.price = price

    def price(self):
        print(f'My price is ${self.price}')

    def model(self):
        print(f'My model is {self.model}')
```
Получается, что класс Model наследует Brand, который наследует AutoModel. Это значит то, что класс потомок имеет доступ к методам родителя и к методам, которые есть у родителя родителя и т.д. Давайте запустим код и проверим!
```
# создадим экземпляр класса Model
raw4 = Model(wheels=4,doors=4,name='Toyota',country='USA'model='raw4', price=15000)
# давайте обратимся к методам и проверим, всё ли работает
raw4.price() # 'My price is $15000'
raw4.sayMyName() #'Im Toyota from USA'
```
Замечательно! Всё работает
